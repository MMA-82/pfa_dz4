## Условие 3: 3 задача

```python
Исследуйте, какие характеристики недвижимости влияют на ее стоимость недвижимости, 
с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: 
ополнить каждый график письменными выводами и наблюдениями.
```

```python
Как известно, цена дома или квартиры напрямую зависит от количества комнат 
в этом самом доме или квартире. Давайте посмотрим насколько эта зависимость достоверна 
в отношении нашей анализируемой базы данных, и отрисуем соответствующий график распределения.
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(houses['bedrooms'], houses['price'])
plt.title('Зависимость цены дома от числа спален')
plt.xlabel('Число спален')
plt.ylabel('Цена дома')
plt.grid()
```

```python
На графике видна максимальная плотность распределения в диапазоне 3-5 спален, 
т.е. дома с таким количеством комнат стоят дороже, а с 5 спальнями самые дорогие.
Также видно, что цена на дома с числом спален больше 5 имеет обратную корреляцию 
- уменьшение стоимости. Следовательно анализа только данной взаимосвязи не достаточно, 
чтобы понять от чего такая цена. 
На стоимость дома влияет также и его возраст. 
Давайте пороанализируем распределение цены домов от года постройки. 
```

```python
plt.figure(figsize=(8, 6))
plt.bar(houses['yr_built'], houses['price'])
plt.title('Зависимость цены дома от года постройки')
plt.xlabel('Год строительства дома')
plt.ylabel('Цена дома')
plt.xticks(rotation = 30)
```

```python
По графику видно, что с наибольшей ценой чаще встречаются дома, 
построенные с конца 90-х по примерно 2010-ые, хотя дома 1920-1940 годов постройки 
также имеют хорошую цену. При том, что совсем новые дома не такие дорогие, 
а в целом держатся в том же диапазоне цены как и более старшие по возрасту дома.
Давайте посмотрим как влияет количество этажей на цену дома.
```

```python
plt.figure(figsize=(8, 6))
plt.plot(houses['floors'], houses['price'])
plt.title('Зависимость цены дома от этажности')
plt.xlabel('Число этажей')
plt.ylabel('Цена дома')
plt.grid()
```
